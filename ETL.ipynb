```python
import pandas as pd
import os 
from tqdm import tqdm_notebook as tqdm
os.listdir('data')
```

```python
def clean_data(df):
    df = df.T.reset_index()
    prefix = df.iloc[1,9:-6].values
    prefix = pd.Series(prefix).fillna(method='ffill').values
    col = df.iloc[2,9:-6].values
    col = [ p+'_'+c for p,c in zip(prefix,col)]
    df = df.iloc[5:,9:-6].reset_index(drop=True)
    df.columns = col
    return df.astype('float64')
```

```python
xls = pd.ExcelFile('./data/前50整合檔.xlsx')
```

```python
df = pd.DataFrame()
for i in tqdm(xls.sheet_names):
    raw = pd.read_excel('./data/前50整合檔.xlsx',sheet_name = i)
    df = df.append(clean_data(raw)).reset_index(drop=True)
    print(i,df.shape)
```
